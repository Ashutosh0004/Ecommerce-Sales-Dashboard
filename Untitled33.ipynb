```python
## 🛒 E-commerce Sales Analysis

## 📌 Problem Statement
#This project analyzes e-commerce sales data to identify key factors affecting revenue and profitability,
 #and provides actionable insights to improve business performance.
```

```python
import pandas as pd
df=pd.read_csv('ecommerce_dirty_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['Sales'].fillna(df['Sales'].mean(),inplace=True)
df.isnull().sum()
```

```python
df['Region'].fillna(df['Region'].mode()[0],inplace=True)
df['Segment'].fillna(df['Segment'].mode()[0],inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
#STEP 2: EDA (Exploratory Data Analysis)
#Sales Trend (Time Analysis)
import matplotlib.pyplot as plt
df.groupby('Order_Date')['Sales'].sum().plot(figsize=(10,5))
plt.title('Sales over Time')
plt.show()

```

```python
#Category Wise Sales
df.groupby('Category')['Sales'].sum().plot(kind='bar',color='red',edgecolor='black')
plt.title('Sales by Category')
plt.show()
```

```python
#Region Wise Sales
df.groupby('Region')['Sales'].sum().plot(kind='bar',color='green',edgecolor='black')
plt.title('Sales by Region')
plt.show()
```

```python
#Profit Analysis
df.groupby('Category')['Profit'].sum().plot(kind='bar',color='orange',edgecolor='black')
plt.title('Category wise Profit')
plt.show()
```

```python
#Discount vs Profit
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='Discount',y='Profit',data=df)
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.show()
```

```python
# Top Customer
df.groupby('Customer_ID')['Sales'].sum()\
  .sort_values(ascending=False)\
  .head(10)\
  .plot(kind='bar')

plt.title('Top Customers')
plt.xlabel('Customer ID')
plt.ylabel('Sales')
plt.show()
```
